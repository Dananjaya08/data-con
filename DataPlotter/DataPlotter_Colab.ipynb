## 1. Upload file ZIP
Upload file `DataPlotter.zip` ke Colab dengan tombol upload di kiri, lalu lanjutkan.

```python
import zipfile, os

zip_path = "/content/DataPlotter.zip"   # lokasi file ZIP yang diupload
extract_path = "/content/DataPlotter"   # folder tujuan ekstrak

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

# cek hasil ekstrak
os.listdir(extract_path)
```
